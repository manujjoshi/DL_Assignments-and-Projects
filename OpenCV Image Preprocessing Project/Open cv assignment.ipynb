```python
import cv2
import matplotlib.pyplot as plt
```

### Read the image

```python
img = cv2.imread('2.jpg')
img.shape
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Display the image

```python
plt.imshow(img)
```

### Rotate the image by 90degrees

```python
rot = img
rot = cv2.rotate(rot, cv2.cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rot)
```

### Crop the image

```python
crop = img
cropped_image = crop[:400,400:800]
plt.imshow(cropped_image)
```

### Resize the image a) reduce the size with row and column length b) with axis values 
[resize cv2](https://www.tutorialkart.com/opencv/python/opencv-python-resize-image/)

### a)

```python
resize = img
resize = cv2.resize(resize,(600,600))
plt.imshow(resize)
```

### b)

```python
resize = img
w_ratio = 0.8
h_ratio = 0.4
resize = cv2.resize(resize,(0,0),resize,w_ratio,h_ratio)
plt.imshow(resize)
```

### Adjust Image Contrast *****
- [Understanding Image Contrast With OpenCV in Python](https://www.analyticsvidhya.com/blog/2021/09/understanding-image-contrast-with-opencv-in-python/)
- [Changing the contrast and brightness of an image using Python  OpenCV](https://www.geeksforgeeks.org/changing-the-contrast-and-brightness-of-an-image-using-python-opencv/)

### Blur the image a) Gaussian b) Median Blur (blur the image by 50%) *****
[Smoothing / Blurring](https://www.geeksforgeeks.org/python-image-blurring-using-opencv/)

```python
blur = img
gaussian_blur = cv2.GaussianBlur(blur,(21,21),8)
plt.imshow(gaussian_blur)
```

```python
med_blur = img
median = cv2.medianBlur(med_blur, 21)
plt.imshow(median)
```

### Detect the edges (Canny Edge Detector) 

```python
canny = img
gra = cv2.cvtColor(canny,cv2.COLOR_RGB2GRAY)

canny_edge_LPF = cv2.Canny(gra,21,21)  # 21,21 is thresholds   # Low Pass filter
plt.imshow(canny_edge_LPF)
print('Canny edge with low pass filter')
plt.show()
print('Canny edge with High pass filter')
canny_edge_HPF = cv2.Canny(gra,172,172)  # High Pass filter
plt.imshow(canny_edge_HPF)
plt.show()
```

### Convert image to grayscale (Black & White)

```python
bw = img
bw = cv2.cvtColor(bw,cv2.COLOR_RGB2GRAY)
plt.imshow(bw)
print('2 dimensions indicate it is a grey scale image',bw.shape)
```
